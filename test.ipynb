```python

def split_codingpos(pos_str:str)->list:
    if len(pos_str.replace(" ", "")) == 0 or pos_str.replace(' ', '') == '0':
        return None
    nested_list  = [[tuple(map(int, p2.split(':'))) for p2 in p1.split(',')] for p1 in pos_str.replace(' ', '').split('|')]
    nested_list = [[(x, y+1) for x, y in inter_tuple] for inter_tuple in nested_list]
    print(f"splitted coding positions is {nested_list}")
    if all(isinstance(item, tuple) and all(isinstance(i, int) for i in item) for sublist in nested_list for item in sublist):
        return nested_list
    else:
        raise ValueError(f"Invalid coding positions string: {pos_str}")

def get_triml_pos(out_lst, trimlpos):
    new_pos = []
    offsetl = 0
    found = False
    if trimlpos <= out_lst[0][0]:
        offsetl = trimlpos + 1
    else:
        for in_lst in out_lst:
            if trimlpos >= in_lst[1]:
                continue
            else:
                for pos in list(range(in_lst[0], in_lst[1])):
                    if pos >= trimlpos:
                        if (pos + 1) % 3 == 0:
                            offsetl = pos + 1
                            found = True
                            break
            if found:
                break
    for in_lst in out_lst:
        if offsetl < in_lst[1]:
            posl = in_lst[0] - offsetl
            if posl < 0:
                posl = 0
            new_pos.append((posl, in_lst[1]-offsetl))
    return new_pos

def get_trimr_pos(out_lst, trimrpos):
    found = False
    offsetr = 0
    for in_lst in out_lst[::-1]:
        for pos in list(range(in_lst[1]-1, in_lst[0]-1, -1)):
            if pos <= trimrpos:
                if pos % 3 == 0:
                    offsetr = pos
                    found = True
                    break
        if found:
            break
    new_lst = []
    if found:
        for in_lst in out_lst:
            if offsetr >= in_lst[1]:
                new_lst.append(in_lst)
            elif in_lst[0] < offsetr < in_lst[1]:
                new_lst.append((in_lst[0], offsetr))
    return new_lst
def get_new_codingpos(codingpos, triml, trimr, dnalen):
    new_codingpos = []
    if triml == 0 and trimr == 0:
        return
    elif triml != 0 and trimr == 0:
        trimlpos = triml - 1
        for out_lst in codingpos:
            if trimlpos < out_lst[-1][-1] - 3:
                new_codingpos.append(get_triml_pos(out_lst, trimlpos))
    elif triml == 0 and trimr != 0:
        trimrpos = dnalen - trimr
        for out_lst in codingpos:
            if trimrpos >= out_lst[-1][1]:
                new_codingpos.append(out_lst)
            else:
                new_codingpos.append(get_trimr_pos(out_lst, trimrpos))
    else:
        trimrpos = dnalen - trimr
        for out_lst in codingpos:
            if trimrpos >= out_lst[-1][1]:
                new_codingpos.append(out_lst)
            else:
                new_codingpos.append(get_trimr_pos(out_lst, trimrpos))
        if len(new_codingpos) != 0:
            trimlpos = triml - 1
            new_new_codingpos = []
            for out_lst in new_codingpos:
                if trimlpos < out_lst[-1][-1] - 3:
                    new_new_codingpos.append(get_triml_pos(out_lst, trimlpos))
            new_codingpos = new_new_codingpos
    return new_codingpos

codingstr = "0:8,15:29|0:8,20:49|0:8,30:89|1:9,100:118"
codingpos = split_codingpos(codingstr)
print(f"codingpos: {codingpos}")

newnew = get_new_codingpos(codingpos, 5, 5, 120)
print(f"newnew: {newnew}")
```
